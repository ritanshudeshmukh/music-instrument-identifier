## Feature Distribution

### Instrument Families

```python
#Display Instrument families names
dfTrain['instrument_family'].value_counts().reindex(np.arange(0,11, 1)).plot(kind = 'bar')
plt.title("Instrument Family Distribution for Training")
plt.xlabel("Instrumet Family")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("Graphs/instrumentsListTraining.jpg")
plt.show()
```

```python
dfValid['instrument_family'].value_counts().reindex(np.arange(0,11, 1)).plot(kind = 'bar')
plt.title("Instrument Family Distribution for Validation")
plt.xlabel("Instrumet Family")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("Graphs/instrumentsListValidation.jpg")
plt.show()
```

```python
dfTest['instrument_family'].value_counts().reindex(np.arange(0,11, 1)).plot(kind = 'bar')
plt.title("Instrument Family Distribution for Test")
plt.xlabel("Instrumet Family")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("Graphs/instrumentsListTest.jpg")
plt.show()
```
